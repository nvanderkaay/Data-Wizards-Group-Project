```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
car_sale = pd.read_csv(r"C:\Users\shayj\Downloads\Online_Cars_Sale_Marketplace (2).csv")
```

```python
car_sale = car_sale[car_sale['Price'] != 'Not Priced'] ## remove 'Not Priced' rows
car_sale['Price'] = car_sale['Price'].replace(r'[,$]', '', regex=True)  ## remove special characters $ and , 
car_sale['Price'] = car_sale['Price'].astype(int) 
car_sale['Used/New'] = car_sale['Used/New'].replace(r'.*Certified.*', 'Certified', regex=True) ## change all Certified to just Certified
car_sale[car_sale['Drivetrain'] != "-"] ## remove "-" in Drivetrain column
Min = car_sale['MinMPG']
Max = car_sale['MaxMPG']
car_sale['MaxMPG'] = np.maximum(Min, Max)
car_sale['MinMPG'] = np.minimum(Min, Max) ## Fix the min and max being in the wrong column
car_sale.drop(car_sale[(car_sale['MinMPG'] == 0)&(car_sale['MaxMPG'] == 0)].index, inplace = True) ## remove rows where min and max MPG are 0 
car_sale = car_sale.drop_duplicates() ## remove duplicates
car_sale            #Andrews clean data, will be using it to try and answer my question
```

```python
car_sale['Age'] = 2025 - car_sale['Year']
#figures out the age of the cars using 2025 as a refrence year

```

```python

car_sale["annual_value_loss"] = car_sale["Price"] / car_sale["Age"]
car_sale["annual_value_loss"]
```

```python
car_sale["annual_value_loss"] = car_sale["annual_value_loss"].round(2)
car_sale["annual_value_loss"]
```

```python

car_sale["annual_value_loss"].head()
```

```python
top_annual_loss = car_sale.sort_values(by="annual_value_loss", ascending=False)
top_5_annual_loss = top_annual_loss.head(5)
top_5_annual_loss
```

As shown here, the top 5 cars with the highest annual loss are very high-end luxury car brands. This trend makes sense, as expensive cars can depreciate pretty heavily, at least in their early years. Since they are higher-quality vehicles, they can lose value a lot faster. With certain luxury car models and brands, the depreciation rate can fluctuate due to things like higher demand, rarity, etc. Generally these cars can lose value faster than normal cars. 

These cars are very new and expensive, which is why the annual value loss is high. They hold a lot of value in a short time because of a higher price and newer vehicle.

lower value car = lower value loss 


higher value car = higher value loss

This annual value loss will go down over time as the car gets older. 

```python
most_depreciated_car = top_5_annual_loss.head(1)
most_depreciated_car
```

```python

plt.figure(figsize=(10, 6))
sns.histplot(car_sale['Price'], binwidth=5000, kde=False, color='green')


plt.title('Distribution of Car Prices', fontsize=16)
plt.xlabel('Price (in dollars)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)


plt.show()
```

This histogram represents the distribution of all the car prices in the dataset. As we can see here, most cars are within the 0 to 100,000 dollar range, and there are some outliers that are hard to see around 200,000, 300,000, and 400,000 dollars. A lot of the cars prices peak near the middle of the 0-100,000 dollar range.
