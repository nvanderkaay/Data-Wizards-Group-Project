```python
import pandas as pd
from google.colab import files
from sklearn.model_selection import train_test_split
from sklearn.linear_model import RidgeCV
from sklearn.metrics import mean_squared_error

#Luxury vecicles by zip code

uploaded = files.upload()

df = pd.read_csv('OLCMP.csv')

luxury_brands = ('Audi', 'BMW', 'Cadillac', 'Mercedes-Benz', 'Porsche', 'Volvo', 'Lexus', 'Lincoln', 'Tesla')

df['LuxuryCar'] = df['Make'].isin(luxury_brands).astype(int)



luxury_sales_by_zip = df[df['LuxuryCar'] == 1].groupby('Zipcode').size()
luxury_sales_by_zip = luxury_sales_by_zip.sort_values(ascending=False)
top_25_zipcodes = luxury_sales_by_zip.head(25)
print(top_25_zipcodes)




```

```python
import matplotlib.pyplot as plt


top_25_zipcodes = top_25_zipcodes.reset_index(name='LuxuryCarSales')



plt.figure(figsize=(12,6))
plt.bar(top_25_zipcodes['Zipcode'].astype(str), top_25_zipcodes['LuxuryCarSales'])
plt.xticks(rotation=90)
plt.xlabel('Zip Code')
plt.ylabel('Number of Luxury Cars Sold')
plt.title('Top 25 Zip Codes for Luxury Car Sales')
plt.tight_layout()
plt.show()
```
