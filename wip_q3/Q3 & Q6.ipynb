```python
#Import pandas for dataframe functionality, data cleaning, methods and tools
import pandas as pd

#Load Dataset
df = pd.read_csv("Online_Cars_Sale_Marketplace.csv")

#View First 20 Rows of Dataset
df.head(20)
```

```python
# Summary of columns, data types, and non-null values
df.info()

# Check for missing values
df.isnull().sum()


```

```python
df["DealType"].unique()
```

```python
# Fill NaN in DealType with a placeholder
df['DealType'] = df['DealType'].fillna('Unknown')

#Look at new DeatType Category "Unknown"
df["DealType"].unique()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
import numpy as np
df['Price'] = df['Price'].replace('Not Priced', np.nan)

#Counting how many "Not Priced" are in "Price" column (should be 0 now)
(df['Price'] == 'Not Priced').sum()


```

```python
#View Rows where Price is NaN
df[df['Price'].isnull()]

```

```python
#clean and convert 'Price' column from object to float
# Step 1: Make sure everything is a string (even NaNs will be converted safely)
df['Price'] = df['Price'].astype(str)

# Step 2: Clean the strings (remove $ and ,), then convert to float
df['Price'] = df['Price'].str.replace(r'[\$,]', '', regex=True)

# Step 3: Convert cleaned strings to float — 'nan' will become actual NaN
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

```

```python
# Confirm "Price" is now a float 
print(df['Price'].dtype)   
```

```python
# Confirm "Price" is now a float 
df.head(20)
```

```python
# Check for fully duplicate rows
duplicate_rows = df[df.duplicated(keep=False)]
print(f"Total duplicate rows: {duplicate_rows.shape[0]}")
display(duplicate_rows)

```

```python
# Identify fully duplicated rows
duplicate_rows = df[df.duplicated(keep=False)]

# Sort the duplicates so they appear grouped together
duplicate_rows = duplicate_rows.sort_values(by=df.columns.tolist())

# Display the count and the duplicates
print(f"Total duplicate rows: {duplicate_rows.shape[0]}")
display(duplicate_rows)
```

```python
# Remove all fully duplicated rows and overwrite the original dataframe
df = df.drop_duplicates()

# Confirm the shape after removing duplicates
print(f"New shape after removing duplicates: {df.shape}")

# Optional: preview cleaned DataFrame
display(df.head())

```

```python
# Step 1: Identify rows where 'Make' is 'Tesla' but 'FuelType' is not 'Electric'
incorrect_fuel_teslas = df[(df['Make'].str.lower() == 'tesla') & (df['FuelType'].str.lower() != 'electric')]

# Step 2: Display the count and the problematic rows
print(f"Total Tesla vehicles with incorrect FuelType: {incorrect_fuel_teslas.shape[0]}")
display(incorrect_fuel_teslas[['Make', 'Model', 'Year', 'FuelType']])

# Step 3: Correct the 'FuelType' for all Tesla vehicles to 'Electric'
df.loc[df['Make'].str.lower() == 'tesla', 'FuelType'] = 'Electric'

# Step 4: Verify the correction
corrected_fuel_teslas = df[df['Make'].str.lower() == 'tesla']
print(f"Total Tesla vehicles after correction: {corrected_fuel_teslas.shape[0]}")
display(corrected_fuel_teslas[['Make', 'Model', 'Year', 'FuelType']].drop_duplicates())
```

```python
# Filter the DataFrame to include only rows where 'Make' is 'Tesla'
tesla_df = df[df['Make'].str.lower() == 'tesla']

# Display the filtered DataFrame
display(tesla_df)
```

```python
# Step 1: Display unique values in the 'Transmission' column
unique_transmissions = df['Transmission'].unique()
print("Unique transmission types in the dataset:")
print(unique_transmissions)

# Step 2: Identify rows with uncommon or possibly incorrect 'Transmission' entries
# Define standard transmission types
standard_transmissions = ['Automatic', 'Manual', 'CVT', 'DCT', 'AMT', 'Semi-Automatic']

# Filter rows with uncommon 'Transmission' entries
uncommon_transmissions_df = df[~df['Transmission'].isin(standard_transmissions)]

# Display the count and the rows with uncommon 'Transmission' entries
print(f"\nTotal rows with uncommon 'Transmission' entries: {uncommon_transmissions_df.shape[0]}")
display(uncommon_transmissions_df[['Transmission', 'Make', 'Model', 'Year']])
```

```python
# Step 1: Filter only Tesla vehicles
tesla_only = df[df['Make'] == 'Tesla']

# Step 2: Select relevant columns - 'Make' and 'Engine'
tesla_engines = tesla_only[['Make', 'Engine']]

# Step 3: Drop duplicates to get unique Make-Engine combinations
tesla_engines_unique = tesla_engines.drop_duplicates()

# Step 4: Display the result
print(tesla_engines_unique)
```

```python
# Step 1: Filter rows where 'Engine' contains 'electric' (case-insensitive)
electric_engine_df = df[df['Engine'].str.contains('electric', case=False, na=False)]

# Step 2: Select and deduplicate relevant columns
electric_engine_df_sample = electric_engine_df[['Make', 'Model', 'Engine']].drop_duplicates()

# Step 3: Display the first 30 rows
print(electric_engine_df_sample.head(30))

```

```python
# Filter for rows where 'Engine' contains 'Electric Motor'
electric_motor_df = df[df['Engine'].str.contains('Electric Motor', case=False, na=False)]

# Select unique Make-Engine combinations
electric_motor_makes = electric_motor_df[['Make', 'Engine']].drop_duplicates()

# Sort and display
print(electric_motor_makes.sort_values(by='Make'))
```
