```python
import pandas as pd
from google.colab import files
from sklearn.model_selection import train_test_split
from sklearn.linear_model import RidgeCV
from sklearn.metrics import mean_squared_error

#Budget vecicles by zip code

uploaded = files.upload()

df = pd.read_csv('OLCMP.csv')

budget_brands = ('Acura', 'Alfa', 'Buick', 'Chevrolet', 'Dodge', 'Ford', 'Genesis', 'GMC', 'Honda', 'Hyundai', 'INFINITI', 'Jeep', 'Kia', 'Mazda','Mercury', 'MINI','Mitsubishi', 'Nissan','RAM', 'Saturn', 'Scion', 'Subaru', 'Toyota', 'Volkswagen')

df['BudgetCar'] = df['Make'].isin(budget_brands).astype(int)



budget_sales_by_zip = df[df['BudgetCar'] == 1].groupby('Zipcode').size()
budget_sales_by_zip = budget_sales_by_zip.sort_values(ascending=False)
top_25_zipcodes = budget_sales_by_zip.head(25)
print(top_25_zipcodes)
```

```python
import matplotlib.pyplot as plt


top_25_zipcodes = top_25_zipcodes.reset_index(name='BudgetCarSales')



plt.figure(figsize=(12,6))
plt.bar(top_25_zipcodes['Zipcode'].astype(str), top_25_zipcodes['BudgetCarSales'])
plt.xticks(rotation=90)
plt.xlabel('Zip Code')
plt.ylabel('Number of Budget Cars Sold')
plt.title('Top 25 Zip Codes for Budget Car Sales')
plt.tight_layout()
plt.show()
```
